#Import data and libraries

```
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://raw.github.sydney.edu.au/awon6941/DATA3406_Group4/master/owid-covid-data.csv?token=AAAA6WPK6RZFDLJJ62KGLJK7SDSXG' 
df = pd.read_csv(url)
```

#Filter data - Japan

```
dfJapan = df[df['location'].str.contains("Japan")]
dfItaly = df[df['location'].str.contains("Italy")]
dfChina = df[df['location'].str.contains("China")]
```

#Exploratory data analysis 

According to [1], Japan has had a surprisingly low death rate compared to other countries. The new cases and deaths per million each day is plotted below for Japan, as well as Italy and China for comparison. 

```
# Plot for Japan 
date = dfJapan['date']
ncase = dfJapan['new_cases_per_million']
ndeath = dfJapan['new_deaths_per_million']

# Adding legend for stack plots is tricky.
plt.plot([], [], color='r', label = 'Confirmed')
plt.plot([], [], color='b', label = 'Deaths')

plt.stackplot(date, ndeath, ncase, colors= ['r', 'b'])
plt.title('Cases in Japan')
plt.ylabel('Cases per million')
plt.legend()
plt.show()

# Plot for Italy
date = dfItaly['date']
ncase = dfItaly['new_cases_per_million']
ndeath = dfItaly['new_deaths_per_million']

plt.plot([], [], color='r', label = 'Confirmed')
plt.plot([], [], color='b', label = 'Deaths')

plt.stackplot(date, ndeath, ncase, colors= ['r', 'b'])
plt.title('Cases in Italy')
plt.ylabel('Cases per million')
plt.legend()
plt.show()

# Plot for China
date = dfChina['date']
ncase = dfChina['new_cases_per_million']
ndeath = dfChina['new_deaths_per_million']

plt.plot([], [], color='r', label = 'Confirmed')
plt.plot([], [], color='b', label = 'Deaths')

plt.stackplot(date, ndeath, ncase, colors= ['r', 'b'])
plt.title('Cases in China')
plt.ylabel('Cases per million')
plt.legend()
plt.show()
```

It can be seen in the plots above that Japan and Italy has seen new cases surge significantly twice since the pandemic began. The proportion of deaths versus new cases in Japan are seem to also be significantly lower than China and Italy. 



```
# plotting total cases in each of these countries 
plt.plot(dfJapan['date'], dfJapan['total_cases'], label="Japan")
plt.plot(dfItaly['date'], dfItaly['total_cases'], label="Italy")
plt.plot(dfChina['date'], dfChina['total_cases'], label="China")
plt.plot()

plt.ylabel("Total cases")
plt.title("Total cases over time")
plt.legend()
plt.show()

# plotting total deaths in each of these countries 
plt.plot(dfJapan['date'], dfJapan['total_deaths'], label="Japan")
plt.plot(dfItaly['date'], dfItaly['total_deaths'], label="Italy")
plt.plot(dfChina['date'], dfChina['total_deaths'], label="China")
plt.plot()

plt.ylabel("Total deaths")
plt.title("Total deaths over time")
plt.legend()
plt.show()
```

As seen in the plots, the total cases in Japan has caught up to the total cases in China at present. However, the total deaths for Japan is significantly lower than both Italy and China.
