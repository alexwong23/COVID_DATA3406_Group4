#Import data

```
import pandas as pd
url = 'https://raw.github.sydney.edu.au/awon6941/DATA3406_Group4/master/owid-covid-data.csv?token=AAAA6WPK6RZFDLJJ62KGLJK7SDSXG' #has to be raw.github link 
df1 = pd.read_csv(url)

print(df1.shape)
```

#Filter data - Italy

```
df1['population (million)'] = df1['population'].astype(float)/1000000

dfItaly = df1[df1['location'].str.contains("Italy")]
dfSing = df1[df1['location'].str.contains("Singapore")]
dfChina = df1[df1['location'].str.contains("China")]

dfItaly
```

#Analysis - Looking at age

```
import pandas as pd
dfAge = pd.DataFrame(df1,columns=['location','median_age','aged_65_older','aged_70_older']) # filter by columns
dfAge = dfAge[dfAge['location'].isin(['Italy', 'Singapore', 'China'])] # filter by country
dfMedianAge = dfAge.groupby(['location'], sort=False)['median_age'].max() # group by max value
df65 = dfAge.groupby(['location'], sort=False)['aged_65_older'].max() # group by max value
df70 = dfAge.groupby(['location'], sort=False)['aged_70_older'].max() # group by max value
print(dfMedianAge)
print(df65)
print(df70)
```

#Analysis - Looking at gdp, hospital beds per thousand, life expectancy

```
import pandas as pd
dfAge = pd.DataFrame(df1,columns=['location','gdp_per_capita','life_expectancy','hospital_beds_per_thousand']) # filter by columns
dfAge = dfAge[dfAge['location'].isin(['Italy', 'Singapore', 'China'])] # filter by country
dfGDP = dfAge.groupby(['location'], sort=False)['gdp_per_capita'].max() # group by max value
dfLife = dfAge.groupby(['location'], sort=False)['life_expectancy'].max() # group by max value
dfBeds = dfAge.groupby(['location'], sort=False)['hospital_beds_per_thousand'].max() # group by max value
print(dfGDP)
print(dfLife)
print(dfBeds)
```

#Analysis - Population

```
import pandas as pd
dfPopulation = pd.DataFrame(df1,columns=['location','population (million)', 'population_density']) # filter by columns
dfPopulation = dfPopulation[dfPopulation['location'].isin(['Italy', 'Singapore', 'China'])] # filter by country
dfMillion = dfPopulation.groupby(['location'], sort=False)['population (million)'].max() # group by max value
dfDensity = dfPopulation.groupby(['location'], sort=False)['population_density'].max() # group by max value
print(dfMillion)
print(dfDensity)

# total population barchart here
```
